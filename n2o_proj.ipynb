```python
years = [2020, 2030, 2040, 2050]
all_yrs = [i for i in range(2020, 2051)]
```

we divide by 64 because to convert from N-equivalence to N2O (N2O is 64% N by mass)

```python
BAU = np.interp(all_yrs, years, [7.0, 8.1, 8.9, 9.7])/0.64
mod_mit = np.interp(all_yrs, years, [6.0, 6.3, 6.5, 6.7])/0.64
con_mit = np.interp(all_yrs, years, [5.2, 5.0, 4.7, 4.2])/0.64
```

```python
plt.plot(all_yrs, BAU, label='BAU')
plt.plot(all_yrs, mod_mit, label='moderate mitigation')
plt.plot(all_yrs, con_mit, label='concerted mitigation')
plt.ylabel('Tg N2O-N/yr')
plt.ylim(0, 16)
plt.legend(loc="center left", bbox_to_anchor=(1, 0, 0.5, 1.7))
plt.show()
```
